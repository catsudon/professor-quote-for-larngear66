```python
import random
from rembg import remove
from PIL import Image, ImageFont, ImageDraw, ImageEnhance, ImageOps

def gen(name, out_dir, student_path):
    bg = Image.open('img/collage.jpg').convert("RGBA"); W, H = bg.size
    student = Image.open(student_path).convert("RGBA")
    student = remove(student)

    # insert student pic
    student = ImageOps.contain(student, bg.size)
    bg.paste(student, ((W-student.size[0])//2, H-student.size[1]), student)

    # insert quote
    img = ImageDraw.Draw(bg)
    mf = ImageFont.truetype('font/MechokhThin.ttf', int(W/11))
    mquote = quote[random.randint(0,len(quote)-1)].format(name=name, nisit="นิสิต") # might change "นิสิต" to "นักเรียน"
    
    # fill shadow
    _, _, w, h = img.textbbox((0, 0), mquote, font=mf)
    x = (W-w)/2
    y = (H-h)/2 + h - h//2
    shadowcolor = "black"
    img.text((x-1, y), mquote, font=mf, fill=shadowcolor) ; img.text((x+1, y), mquote, font=mf, fill=shadowcolor) ; img.text((x, y-1), mquote, font=mf, fill=shadowcolor) ; img.text((x, y+1), mquote, font=mf, fill=shadowcolor)
    img.text((x+1, y-1), mquote, font=mf, fill=shadowcolor) ; img.text((x-1, y-1), mquote, font=mf, fill=shadowcolor) ;img.text((x-1, y+1), mquote, font=mf, fill=shadowcolor) ; img.text((x+1, y+1), mquote, font=mf, fill=shadowcolor)

    img.text((x, y), mquote, (201,201,28), font=mf)
    
    # bg.show()
    bg.save(out_dir)
```

```python
for i in range(10):
    gen("การ์ฟิล", "dist/out{x}.png".format(x=str(i)), "./img/student/01.jpg")
```
